```python
class Book:
    def __init__(self, title, author, isbn, date):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.date = date

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "978-0743273565", "1925")
book2 = Book("1984", "George Orwell", "978-0451524935", "1949")
book3 = Book("To Kill a Mockingbird", "Harper Lee", "978-0061120084", "1960")
book4 = Book("Moby-Dick", "Herman Melville", "978-1503280786", "1851")
book5 = Book("The Catcher in the Rye", "J.D. Salinger", "978-0316769488", "1951")

name = "Luca Library"

books = [book1, book2, book3, book4, book5]
lent_books = []

def display_books():
    print(f"Welcome to {name}! Here are the books available in the library:")
    for d in books:
        print(f"  {d.title} by {d.author} ISBN: {d.isbn}, Year: {d.date})")

class User:
    def __init__(self, name, phone, email, password):
        self.name = name
        self.phone = phone
        self.email = email
        self.password = password

def register_user():
    print("Register now to use the library system!")
    register_name = input("Enter your full name: ")
    register_number = input("Enter your phone number: ")
    register_email = input("Enter your Gmail: ")
    register_password = input("Enter your password: ")

    user = User(register_name, register_number, register_email, register_password)
    print(f"Registration successful! Welcome {register_name}.")

    return user

def lend_book(user):
    display_books()
    booklend = input("Enter the title of the book you want to lend: ")

    for i in books:
        if i.title.lower() == booklend.lower():
            lent_books.append(i)  
            books.remove(i)  
            print(f"The book is all your.")
            return
            
    print(f"Sorry, the book '{booklend}' was not found in the library.")

def main():
    print(f"Welcome to {name}!")
    user = None  

    while True:
        print("Please choose an option:")
        print("1. Register")
        print("2. Show me The books you have")
        print("3. Lend a book")
        print("4. Show lent books")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            user = register_user()
            

        elif choice == "2":
            display_books()

        elif choice == "3":
            if user:
                lend_book(user)
            else:
                print("Sorry, you need to register first.")

        elif choice == "4":
            if lent_books:
                print("Lent books:")
                for book in lent_books:
                    print(f" - {book.title} by {book.author} (ISBN: {book.isbn}, Year: {book.date})")
            else:
                print("No books have been lent out.")

        elif choice == "5":
            break

        else:
            print("Invalid option! Please choose a valid option.")

if __name__ == "__main__":
    main()
```
